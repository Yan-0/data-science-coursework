```python
plt.figure()

#Scatter plot with bill against tip
plt.scatter(data['total_bill'], data['tip'])

#Adding title to plot
plt.title("Total Bill vs Tip")

plt.xlabel("Total Bill")
plt.ylabel("Tip")

plt.show()
```

### Line Chart

```python
x = np.linspace(0,20)
y1 = np.sin(x)
y2 = np.cos(x)

#plot
plt.figure()

#line chart
plt.plot(x, y1)
plt.plot(x, y2)
plt.show()
```

```python
plt.figure()
plt.plot(data['tip'], color="gray")
plt.title("Tip line chart")
plt.xlabel("Index")
plt.ylabel("Tip")
plt.show()
```

### Bar Chart

```python
#Comparing categories OR showing temporal trends

years = np.arange(2012, 2015)
values = [2, 5, 9]
plt.figure()
plt.bar(years, values)
plt.show()
```

```python
years = np.arange(2012, 2015)
category1_values = [2, 5, 9]
category2_values = [7, 6, 3]

plt.figure()

plt.bar(years - 0.2,
        category1_values,
        color = "blue",
        edgecolor = 'none',
        width = 0.4,
        align = 'center',
        label = 'Male'
       )
plt.bar(years + 0.2,
        category2_values,
        color = "yellow",
        edgecolor = 'none',
        width = 0.4,
        align = 'center',
        label = 'Female'
       )

plt.xticks(years, [str(year) for year in years])

plt.legend()

plt.show()
```

```python
plt.figure()

plt.bar(data['day'], data['tip'], color='green')
plt.show()
```

### Pie Chart

```python
#Displaying single proportion

counts = [17, 14]
plt.figure(figsize = (4, 4))

plt.pie(counts,
       colors=['blue', 'red'],
        labels=['Category A', 'Other Categories'],
        startangle=90,
        autopct = '%1.3f%%' #decimal value
       )
plt.show()

#explode parameter
```

### Histogram

```python
#Showing the spread of data column

column_data = np.random.normal(40, 3, 1000)

plt.figure()

plt.hist(column_data,
        color="red",
        bins=50)
plt.show()
```

```python
#Now with our dataframe

plt.hist(data['total_bill'], bins=30) #bins specifies data spread and preciseness

plt.title('Histogram')

plt.show()
```
