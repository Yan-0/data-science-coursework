```python
car_data_null['fueltype'].isna().sum()
```

#### option 1

```python
car_data_null['fuel_type_filled'] = car_data_null['fueltype'].fillna('gas')
```

```python
car_data_null[car_data_null['fuel_type_filled'] == 'gas']
```

#### option 2

```python
car_data_null['fueltype'].fillna('Shreyan', inplace=True)
```

```python
car_data_null[car_data_null['fueltype'] == 'Shreyan']
```

```python
car_data_null['enginelocation'].fillna(car_data_null['enginelocation'].mode()[0], inplace=True)
```

```python
car_data_null['enginelocation'].isna().sum()
```

#### Replacing columns values

```python
car_data_null['fueltype'] = car_data_null['fueltype'].replace({'Shreyan':'gas'})
```

## Data Preprocessing (Categorical to Numerical)

```python
from sklearn.preprocessing import LabelEncoder
encoder = LabelEncoder()
x = ["A", "B"]
x_encoded = encoder.fit_transform(x)
print(x_encoded)
```

```python
car_data['enginetype_encoded'] = encoder.fit_transform(car_data['enginetype'])
```

```python
car_data[['enginetype_encoded', 'enginetype']]
```

```python
car_data['fueltype_enc'] = encoder.fit_transform(car_data['fueltype'])
```

```python
car_data[['fueltype_enc', 'fueltype']]
```
